# Hypothesis testing

With hypothesis testing we want to understand if there are significant differences between different atrributes in this dataset. We will examine the dataset regarding to Location, Titles, Salaries and Job Count. 

```python
def one_sided_test(x):
    pval =  x['P-value']/2
    stats = x['Stats']
    if (pval < 0.05) and (stats > 0):
        one_sided_greater = True
    else:
        one_sided_greater = False
    if (pval < 0.05) and (stats < 0):
         one_sided_smaller = True
    else:
         one_sided_smaller = False
    return one_sided_greater, one_sided_smaller
```

## Location

### Q1: Do different locations hire different proportion of specific roles than others? 

H0: There is no difference in the proportion in the titles hired between different locations.  
H1: There is a difference of a certain region and the mean. 

```python
# Iterate through each selected titles

selected_titles = ['Data Engineer', 'Data Scientist', 'Data Analyst', 'Business Intelligence Consultant/Engineer', 'Business Analyst']

df_prop_test = pd.DataFrame(columns=['Location', 'Title', 'Stats', 'P-value', 'Title Count', 'Nobs', 'H0 Rejected' ])


for title in selected_titles:
    for loc in top_locs:
        # remove LOCATION FILTER FOR POPULATION
        title_count_pop = jobs['Title Grouped'][jobs['Title Grouped']==title].count()
        nobs_pop = jobs['Title Grouped'].count()
        pop_p = title_count_pop/nobs_pop 
    
        count_loc = jobs['Title Grouped'][(jobs['Title Grouped']==title) & (jobs['Location Grouped'] == loc)].count()
        nobs_loc = jobs['Title Grouped'][jobs['Location Grouped'] == loc].count()
        

        
        stats, pval = sm.stats.proportions_ztest(count_loc, nobs=nobs_loc, value=pop_p)
        
        h0 = pval < 0.05
        
        df_prop_test = df_prop_test.append({'Location':loc, 'Title': title,'Stats': stats, 'P-value': pval,'Title Count':count_loc, 'Nobs': nobs_loc, 'H0 Rejected': h0}, ignore_index=True)
        
    
        if pval < 0.05:
            print(f"The H0: {loc} hires the same proportion of {title}'s than the sample: t-stat {round(stats,2)}, p-value {pval}. The H0 can be rejected:", pval < 0.05)
df_prop_test[df_prop_test['H0 Rejected']==True]
```

```python
df_prop_test['One sided Greater'] = df_prop_test.apply(lambda x: one_sided_test(x)[0], axis=1)
df_prop_test['One sided Smaller'] = df_prop_test.apply(lambda x: one_sided_test(x)[1], axis=1)
```

### Q2: Do different locations pay more to specific titles than in other locations?

H0: There is no difference in the mean salaries in the titles between different locations.  
H1: There is a difference of a certain region and the mean.

```python

# Iterate through each selected titles

# Two-tailed test

from scipy.stats import ttest_ind

salary_columns = ['Normalized Sal Min', 'Normalized Sal Max', 'Normalized Sal Avg']


df_t_sal = pd.DataFrame(columns=['Location', 'Title', 'Salary Type', 'Stats', 'P-value', 'Sample', 'H0 Rejected'])



for title in selected_titles:
    for loc in top_locs:
        for sal in salary_columns:
            sal_title = jobs[sal][(salaries['Title Grouped']==title) & (jobs['Location Grouped'] != loc)]
            sal_title_loc = jobs[sal][(salaries['Title Grouped']==title) & (jobs['Location Grouped'] == loc)]
        
        
            n = len(sal_title_loc)
            
# H0 there is no difference between the groups 
# H1 there is a difference


            stats, pval = ttest_ind(sal_title_loc, sal_title, equal_var=False)
            h0 = pval < 0.05
    
            df_t_sal = df_t_sal.append({'Location':loc, 'Title':title, 'Salary Type':sal, 'Stats':stats, 'P-value':pval, 'Sample':n, 'H0 Rejected': h0}, ignore_index=True)             
    
            if pval < 0.05:
                print(f"\n H0: {loc} has a different mean in {sal} for {title}'s than the rest of sample: t-stat {round(stats,2)}, p-value {pval}. The H0 can be rejected:", pval < 0.05)
```

```python
df_t_sal['One sided Greater'] = df_t_sal.apply(lambda x: one_sided_test(x)[0], axis=1)
df_t_sal['One sided Smaller'] = df_t_sal.apply(lambda x: one_sided_test(x)[1], axis=1)
```
