#### Correlation Matrix

```python
cdd.corr().min()
```

```python
# Drop na 
fig, ax = plt.subplots(figsize=(15,10)) 
ax = sns.heatmap(cdd.corr(), annot=True, linewidths=0.5, ax=ax)
bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5) 

```

Find 3 pairs of columns, one pair that is highly positively correlated, one for which the correlation is almost zero and one pair that is negatively correlated. In each case write down an explanation why we see this correlation

```python
# First pair - High correlation 
# Injured/Infected and municipal cost - corr 0.97  
print("The correlation between municipal cost and injured/infected is high, as it is natural that the more injuries there are \n the higher the cost for the municipality is.")
```

```python
# Second pair - Almost zero correlation
# Fatalities and Evacuated correlation -0.0035 
print("There is almost no correlation between fatalities and evacuated. This is interesting as we would maybe expect the fatalities to decrease when evacuation goes up.")
```

```python
# Third paid - Negative correlation 
# There is not really a highyl negatively correlated values. The most negatively correlated is ODG Costs - Utility-People Affected which is -.093
# This means almost no correlation 

print("There are no strongly negatively correlated attributes in this dataset. This makes sense as the type of data we are looking at \nis costs and number of people affected, hence naturally the negative correlations are very slight and not significant.")


```

Plotting the pairs

```python
# First Pair 
sns.pairplot(cdd, vars=["INJURED / INFECTED", "MUNICIPAL COSTS"], kind='reg')
```

The regression line helps us identify that there is a trend and positive correlation. However, we should check the data points more carefully the look for outliers and null values to draw better conclusions. 

```python
sns.pairplot(cdd, vars=["FATALITIES", "EVACUATED"], kind='reg')
```

As our regression trend line shows here it's almost straight without a slope which mean that the data is spread out accross the plot and does not form a correlation

```python
sns.pairplot(cdd, vars=["OGD COSTS", "UTILITY - PEOPLE AFFECTED"], kind='reg')
```

Here we see a light negative slope, but as mentioned before the slope is so small that it basically means no correlation.
