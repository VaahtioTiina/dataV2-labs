### Implementation: Shuffle and Split Data
Split the data into the testing and training datasets. Shuffle the data as well to remove any bias in selecting the traing and test. 

```python
# Your code here
from sklearn.model_selection import train_test_split
```

```python
y = boston['medv']
X = boston.drop(columns={'medv'})

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

## Analyzing Model Performance
Next, we are going to build a Random Forest Regressor, and test its performance with several different parameter settings.

### Learning Curves
Lets build the different models. Set the max_depth parameter to 2, 4, 6, 8, and 10 respectively. 

```python
# Five separate RFR here with the given max depths 
from sklearn.ensemble import RandomForestRegressor
```

```python
regr_2 = RandomForestRegressor(max_depth=2, random_state=0)
regr_2.fit(X_train, y_train)

regr_4 = RandomForestRegressor(max_depth=4, random_state=0)
regr_4.fit(X_train, y_train)

regr_6 = RandomForestRegressor(max_depth=6, random_state=0)
regr_6.fit(X_train, y_train)

regr_8 = RandomForestRegressor(max_depth=8, random_state=0)
regr_8.fit(X_train, y_train)

regr_10 = RandomForestRegressor(max_depth=10, random_state=0)
regr_10.fit(X_train, y_train)
```

Now, plot the score for each tree on the training set and on the testing set.

```python
# Produce a plot with the score for the testing and training for the different max depths

performance_df = pd.DataFrame(columns=['depth', 'r2_score', 'train_test'])


regr2_y_train_pred = regr_2.predict(X_train)
regr2_y_test_pred = regr_2.predict(X_test)
regr2_r2 = r2_score(y_train, regr2_y_train_pred)
regr2_r22 = r2_score(y_test, regr2_y_test_pred) 


regr4_y_train_pred = regr_4.predict(X_train)
regr4_y_test_pred = regr_4.predict(X_test)
regr4_r2 = r2_score(y_train, regr4_y_train_pred)
regr4_r22 = r2_score(y_test, regr4_y_test_pred) 

regr6_y_train_pred = regr_6.predict(X_train)
regr6_y_test_pred = regr_6.predict(X_test)
regr6_r2 = r2_score(y_train, regr6_y_train_pred)
regr6_r22 = r2_score(y_test, regr6_y_test_pred) 

regr8_y_train_pred = regr_8.predict(X_train)
regr8_y_test_pred = regr_8.predict(X_test)
regr8_r2 = r2_score(y_train, regr8_y_train_pred)
regr8_r22 = r2_score(y_test, regr8_y_test_pred) 

regr10_y_train_pred = regr_10.predict(X_train)
regr10_y_test_pred = regr_10.predict(X_test)
regr10_r2 = r2_score(y_train, regr10_y_train_pred)
regr10_r22 = r2_score(y_test, regr10_y_test_pred) 

train_r2 = [regr2_r2, regr4_r2, regr6_r2, regr8_r2, regr10_r2]
test_r2 = [regr2_r22, regr4_r22, regr6_r22, regr8_r22, regr10_r22]
depths = [2, 4, 6,8,10]

r2_df = pd.DataFrame({'depths':depths, 'train_r2': train_r2, 'test_r2':test_r2})

plt.plot(r2_df['depths'], r2_df['train_r2'], label='train R2')
plt.plot(r2_df['depths'], r2_df['test_r2'], label='test R2')
plt.legend()
plt.xlabel('Three depth')
plt.ylabel('R2 score')
plt.title('Model performance of train and test data')
```
